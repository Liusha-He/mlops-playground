```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("seaborn")
```

```python
stock2index = {
    "Tesco": "TSCO.L",
    "HSBC": "HSBC",
    "Microsoft": "MSFT",
    "Netflix": "NFLX",
    "Google": "GOOG",
    "BP": "BP",
    "EasyJet": "EZJ.L",
    "Aviva": "AV.L",
    "Legal&General": "LGEN.L",
    "IBM": "IBM",
    "Fairfax": "FFH.TO",
    "Blackstone": "BX",
    "LloydsBank": "LYG",
    "JPMorgan": "JPM"
}
```

```python
stock_list = [v for v in stock2index.values()]

df = yf.download(stock_list, "2016-01-01")
```

```python
returns = df["Adj Close"].pct_change()
returns = returns.dropna()

returns.head(15)
```

```python
td_year = (returns.count() / ((returns.index[-1] - returns.index[0]).days / 365.25)).iloc[0]

td_year
```

```python
TMR = 0   # target mean returns, assuming no negative returns
```

```python
excess_returns = returns - TMR

excess_returns.head(15)
```

```python
excess_returns = excess_returns.apply(lambda r: np.where(r < 0, r, 0))

excess_returns
```

```python
downsize_deviations = excess_returns.apply(lambda r: np.sqrt(np.mean(r**2)))

downsize_deviations
```

## Sortino Ratio

```python
mean_returns = returns.mean()

mean_returns
```

```python
sortino = (mean_returns - TMR) / downsize_deviations * np.sqrt(td_year)
```

```python
sortino
```

```python
summary = pd.DataFrame(data={"returns": mean_returns, 
                             "risk": downsize_deviations, 
                            "sortino": sortino})

summary
```

```python
summary[summary["sortino"] > 1].sort_values(by=["sortino"], ascending=False)
```
