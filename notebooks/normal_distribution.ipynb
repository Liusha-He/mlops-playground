```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
mu = 0
sigma1 = 1
sigma2 = 2
sigma3 = 3

dx = np.linspace(-7, 7, 1_000)

p1 = generate_pdf(dx, mu, sigma1)
p2 = generate_pdf(dx, mu, sigma2)
p3 = generate_pdf(dx, mu, sigma3)
```
