```python
from datetime import datetime

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
import yfinance as yf
```

```python
df = pd.read_csv("../../data/tutorials/airline_passengers.csv", index_col=["Month"], parse_dates=True).dropna()
```

```python
df.head()
```

```python
df.plot()
```

```python
results = seasonal_decompose(df, model="multiplicative")
```

```python
results.plot()
plt.show()
```

```python
sp_df = yf.download('^GSPC', start=datetime(2000, 1, 1)).dropna()
```

```python
sp_df = sp_df.resample("M").mean()
```

```python
sp_df.head()
```

```python
results_sp = seasonal_decompose(
    sp_df["Adj Close"], 
    model="multiplicative"
)
```

```python
results_sp.trend.plot()
plt.show()
```
